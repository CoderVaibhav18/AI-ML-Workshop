```python
import matplotlib.pyplot as plt
import numpy as np

```

### X Axis = Horizontal Axis,  Y Axis = Vertical Axis


```python
a = np.array([0, 7])
b = np.array([0, 150])
plt.plot(a, b)
plt.show()
```

```python
a = np.array([0, 7, 10, 4]) 
b = np.array([0, 150, 40, 50])
plt.plot(a, b)
plt.show()
```

```python
a = np.array([0, 7, 10, 4]) 
b = np.array([0, 150, 40, 50])
plt.plot(a, b, ".")
plt.show()
```

```python
a = np.array([0, 7, 10, 4]) 
b = np.array([0, 150, 40, 50])
plt.plot(a, b, ".")
plt.show()
```

```python
a = np.array([0, 7, 10, 4]) 
b = np.array([0, 150, 40, 50])
plt.plot(a, b, "o")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
a = np.array([0, 7, 10, 4]) 
b = np.array([0, 150, 40, 50])
plt.plot(a, b, "o")
plt.title("My DM")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
a = np.array([0, 2, 5, 8, 9, 12]) 
b = np.array([0, 40, 50, 80,90,120])
plt.plot(a, b, marker="*")
plt.plot(a, b, marker="*")
plt.plot(a, b, marker="*")
plt.plot(a, b, marker="*")
plt.plot(a, b, marker="*")
plt.plot(a, b, marker="*")
plt.plot(a, b, marker="*")
plt.title("My DM")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
a = np.array([0, 2, 5, 8, 9, 12]) 
b = np.array([0, 40, 50, 80,90,120])
plt.plot(a, b, "o:")
plt.title("My DM")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
a = np.array([0, 2, 5, 8, 9, 12]) 
b = np.array([0, 40, 50, 80,90,120])
plt.plot(a, b, "o:c")
plt.title("My DM")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
a = np.array([0, 2, 5, 8, 9, 12]) 
b = np.array([0, 40, 50, 80,90,120])
plt.plot(a, b, "o--c", ms=10, mec='b', mfc='m')
plt.title("My DM")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python

a = np.array([0, 2, 5, 8, 9, 12]) 
b = np.array([0, 40, 50, 80,90,120])
plt.plot(a, b, "o--c", ms=10, mec='black', mfc='pink')
plt.title("My DM")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid()
plt.show()
```
